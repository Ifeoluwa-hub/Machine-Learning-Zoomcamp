```python
# 7a. Select all the lotus cars
lotus_cars = data[data['Make']== 'Lotus']
lotus_cars
```

```python
# 7b. Select only columns "Engine HP", "Engine Cylinders"
two_columns = lotus_cars[["Engine HP", "Engine Cylinders"]]
two_columns
```

```python
# 7c. Drop all duplicated rows
two_columns.drop_duplicates()
```

```python
# 7d. Get the underlying NumPy array
X = np.array(two_columns)
X
```

```python
# 7e. Compute matrix-matrix multiplication between the transpose of X and X
XTX = np.dot(X.T, X)
XTX
```

```python
#7F. Compute the inverse of XTX
inv = np.linalg.inv(XTX)
```

```python
# 7g. Create an array y with values [1100, 800, 750, 850, 1300, 1000, 1000, 1300, 800]
values = [1100, 800, 750, 850, 1300, 1000, 1000, 1300, 800]
y = np.array(values)
```

```python
# 7i. Multiply the inverse of XTX with the transpose of X, and then multiply the result by y. Call the result w.
result1 = np.dot(inv, X.T)
w = np.dot(result1, y)
```
